```python
import numpy as np

data = np.genfromtxt("deviga.csv", delimiter=",", skip_header=1, dtype=str)
```

```python
print(data)
```

```python
user_id = data[:,0].astype(int)
page = data[:,1]
device = data[:,2]
session = data[:,3].astype(int)
clicks = data[:,4].astype(int)
scroll = data[:,5].astype(int)
bounce = data[:,6].astype(int)
conversion = data[:,7].astype(int)
```

```python
print("Total Sessions:", len(user_id))
```

```python
print("Avg Session Time:", np.mean(session))
```

```python
print("Max Session:", np.max(session))
print("Min Session:", np.min(session))
```

```python
print("Total Conversions:", np.sum(conversion))
```

```python
conv_rate = (np.sum(conversion) / len(conversion)) * 100
print("Conversion Rate:", conv_rate)
```
